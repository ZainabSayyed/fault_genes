```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import svm
from sklearn.metrics import classification_report
import pandas as pd
import sys
import os
import time
df_NOVA=pd.read_csv("D:\\OpenStack\\df.csv")
df_NOVA=df_NOVA.dropna()
#df.take(np.random.permutation(len(df))[:2])
train_NOVA = df_NOVA.sample(frac=0.7)
test_NOVA = df_NOVA.loc[~df_NOVA.index.isin(train_NOVA.index)]
#train_NOVA=train_NOVA.reset_index()
#test_NOVA=test_NOVA.reset_index()
train_data_NOVA=train_NOVA.ix[:,8]
train_labels_NOVA=train_NOVA.ix[:,9]
test_data_NOVA=test_NOVA.ix[:,8]
test_labels_NOVA=test_NOVA.ix[:,9]
# Create feature vectors,  
#x = v.fit_transform(df['Review'].values.astype('U'))  ##
vectorizer = TfidfVectorizer(min_df=5,max_df = 0.8,sublinear_tf=True,use_idf=True)
train_vectors = vectorizer.fit_transform(train_data_NOVA)
test_vectors = vectorizer.transform(test_data_NOVA)

# Perform classification with SVM, kernel=rbf
classifier_rbf = svm.SVC()
t0 = time.time()
classifier_rbf.fit(train_vectors, train_labels_NOVA)
t1 = time.time()
prediction_rbf = classifier_rbf.predict(test_vectors)
t2 = time.time()
time_rbf_train = t1-t0
time_rbf_predict = t2-t1

# Perform classification with SVM, kernel=linear
classifier_linear = svm.SVC(kernel='linear')
t0 = time.time()
classifier_linear.fit(train_vectors, train_labels_NOVA)
t1 = time.time()
prediction_linear = classifier_linear.predict(test_vectors)
t2 = time.time()
time_linear_train = t1-t0
time_linear_predict = t2-t1

# Perform classification with SVM, kernel=linear
classifier_liblinear = svm.LinearSVC()
t0 = time.time()
classifier_liblinear.fit(train_vectors, train_labels_NOVA)
t1 = time.time()
prediction_liblinear = classifier_liblinear.predict(test_vectors)
t2 = time.time()
time_liblinear_train = t1-t0
time_liblinear_predict = t2-t1

# Print results in a nice table
print("Results for SVC(kernel=rbf)")
print("Training time: %fs; Prediction time: %fs" % (time_rbf_train, time_rbf_predict))
print(classification_report(test_labels_NOVA, prediction_rbf))
print("Results for SVC(kernel=linear)")
print("Training time: %fs; Prediction time: %fs" % (time_linear_train, time_linear_predict))
print(classification_report(test_labels_NOVA, prediction_linear))
print("Results for LinearSVC()")
print("Training time: %fs; Prediction time: %fs" % (time_liblinear_train, time_liblinear_predict))
print(classification_report(test_labels_NOVA, prediction_liblinear))
```
